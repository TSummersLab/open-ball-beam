```python
import json

import pandas as pd
import plotly.express as px

from ballbeam.paths import ROOT_PATH
```

```python
EXPERIMENT_NAME = "random"
# EXPERIMENT_NAME = "random_waves_with_inactive"
EXPERIMENT_DIR_PATH = ROOT_PATH / "data" / EXPERIMENT_NAME
EXPERIMENT_TRUNCATED_DIR_PATH = EXPERIMENT_DIR_PATH / "truncated"
```

```python
stems = [p.stem for p in sorted(EXPERIMENT_DIR_PATH.glob("*.json"))]
```

```python
N_warmup = 500
N_clean = 2500
```

```python
def to_json_dict(df):
    return list(df.to_dict(orient="index").values())
```

```python
for stem in stems:
    path = ROOT_PATH / "data" / EXPERIMENT_NAME / f"{stem}.json"
    with path.open() as f:
        log_data = json.load(f)
    df = pd.DataFrame(log_data)

    df = df.iloc[N_warmup:N_warmup+N_clean]
    
    # Write the truncated dataframe back out to JSON
    out_json_data = to_json_dict(df)
    filename = stem + "_truncated" + ".json"
    out_path = EXPERIMENT_TRUNCATED_DIR_PATH / filename
    with out_path.open("w") as f:
        json.dump(out_json_data, f, indent=4)
```

```python
def load(path):
    with path.open() as f:
        return pd.DataFrame(json.load(f))

df_map = {path.stem: load(path) for path in EXPERIMENT_TRUNCATED_DIR_PATH.glob("*.json")}
dfs = list(df_map.values())
```

```python
px.line(dfs[0], y=["action", "observation"])
```
